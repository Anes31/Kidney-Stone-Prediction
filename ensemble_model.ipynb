```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from lightgbm import LGBMClassifier

from pycaret.classification import setup, evaluate_model, compare_models, plot_model

import optuna
```

# Load Data

```python
train = pd.read_csv('train.csv', index_col='id')
test = pd.read_csv('test.csv', index_col='id')
```

```python
X = train.drop(['gravity', 'ph', 'osmo', 'urea', 'target'], axis=1).copy()
y = train.target
```

# Base Model

```python
rf = RandomForestClassifier(random_state=0)

results = []

kf = KFold(n_splits=10, shuffle=True, random_state=0)

r = cross_val_score(rf, X, y, scoring='roc_auc')

results.append(r)

print(f'ROC score: {np.mean(results).round(2)}')
```

# Hyperparameters Tuning

```python
def rf(trial):
    max_depth = trial.suggest_int('max_depth', 1, 100)
    n_estimators = trial.suggest_int('n_estimators', 10, 500)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 100)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 100)
          
    model = RandomForestClassifier(
        max_depth=max_depth,
        n_estimators=n_estimators,
        min_samples_leaf=min_samples_leaf,
        min_samples_split=min_samples_split
    )
    
    model.fit(X, y)
    
    cv_scores = cross_val_score(model, X, y, scoring='roc_auc', cv=kf)
    
    return np.mean(cv_scores)
```

```python
def lr(trial):
    C = trial.suggest_loguniform('C', 0.001, 1000)
    solver = trial.suggest_categorical('solver', ['newton-cg', 'liblinear', 'sag', 'saga'])
    
    if solver=='sag' or solver=='newton-cg':
        penalty = 'l2'
        multi_class = trial.suggest_categorical('multi_class', ['ovr', 'multinomial'])
    elif solver=='liblinear':
        multi_class = 'ovr'
        penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])
        if penalty==1:
            l1_ratio = trial.suggest_float(0, 1)
    elif solver=='saga':
        multi_class = trial.suggest_categorical('multi_class', ['ovr', 'multinomial'])
        penalty = trial.suggest_categorical('penalty', ['l1', 'l2'])
        if penalty==1:
            l1_ratio = trial.suggest_float(0, 1)
          
    model = LogisticRegression(
        C=C,
        penalty = penalty,
        solver=solver,
        multi_class=multi_class
    )
    
    model.fit(X, y)
    
    cv_scores = cross_val_score(model, X, y, scoring='roc_auc', cv=kf)
    
    return np.mean(cv_scores)
```

```python
def lda(trial):
    solver = trial.suggest_categorical('solver', ['svd', 'lsqr', 'eigen'])
          
    model = LinearDiscriminantAnalysis(
        solver=solver
    )
    
    model.fit(X, y)
    
    cv_scores = cross_val_score(model, X, y, scoring='roc_auc', cv=kf)
    
    return np.mean(cv_scores)
```

```python
def et(trial):
    max_depth = trial.suggest_int('max_depth', 1, 100)
    n_estimators = trial.suggest_int('n_estimators', 10, 500)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 100)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 100)
    max_features = trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2', None])

          
    model = ExtraTreesClassifier(
        max_depth=max_depth,
        n_estimators=n_estimators,
        min_samples_leaf=min_samples_leaf,
        min_samples_split=min_samples_split,
        max_features=max_features
    )
    
    model.fit(X, y)
    
    cv_scores = cross_val_score(model, X, y, scoring='roc_auc', cv=kf)
    
    return np.mean(cv_scores)
```

```python
def lightgbm(trial):
    params = {
        'objective': 'binary',
        'boosting_type': 'gbdt',
        'metric': 'binary_logloss',
        'n_estimators': trial.suggest_int('n_estimators', 10, 500),
        'max_depth': trial.suggest_int('max_depth', 1, 50),
        'num_leaves': trial.suggest_int('num_leaves', 2, 100),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.1),
        'min_child_samples': trial.suggest_int('min_child_samples', 1, 100),
        'subsample': trial.suggest_uniform('subsample', 0.1, 1),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.1, 1),
        'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-9, 10.0),
        'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-9, 10.0),
        'random_state': 0
    }

    model = LGBMClassifier(**params)

    model.fit(X, y)
    
    cv_scores = cross_val_score(model, X, y, scoring='roc_auc', cv=kf)
    
    return np.mean(cv_scores)
```

```python
rf_params = {
    'max_depth': 57,
    'n_estimators': 109,
    'min_samples_leaf': 6,
    'min_samples_split': 63
}

lr_params = {
    'C': 0.012026718021443332, 
    'solver': 
    'liblinear', 
    'penalty': 'l2'
}

et_params = {
    'max_depth': 34,
    'n_estimators': 455,
    'min_samples_leaf': 14,
    'min_samples_split': 84,
    'max_features': None
}

lda_params = {
    'solver': 'svd'
}

lightgbm_params = {
    'n_estimators': 357,
    'max_depth': 44,
    'num_leaves': 70,
    'learning_rate': 0.022549425030014225,
    'min_child_samples': 100,
    'subsample': 0.5979184580670622,
    'colsample_bytree': 0.6181123968265769,
    'reg_alpha': 5.490331894344799e-07,
    'reg_lambda': 4.184403255225925e-09
}
```

# Bagging Ensemble

```python
rf_model = RandomForestClassifier(**rf_params)
lda_model = LinearDiscriminantAnalysis(**lda_params)
lr_model = LogisticRegression(**lr_params)
et_model = ExtraTreesClassifier(**et_params)
lightgbm_model = LGBMClassifier(**lightgbm_params)
```

```python
final_model = VotingClassifier(estimators=[('rf', rf_model),
                                           ('lda', lda_model),
                                           ('et', et_model)], 
                                 voting='soft')
```

# Evaluate

```python
results_ensemble = []

r_ensemble = cross_val_score(final_model, X, y, scoring='accuracy')
results_ensemble.append(r_ensemble)
print(f'Accuracy: {np.mean(results_ensemble).round(4)}')
```
